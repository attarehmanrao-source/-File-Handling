### 1Ô∏è‚É£ File Create + Write

```python
f = open("data.txt", "w")   # create + write mode
f.write("Hello, this is first line\n")
f.write("Python file handling start\n")
f.close()
```

### 2Ô∏è‚É£ Read File

```python
f = open("data.txt", "r")
print(f.read())
f.close()
```

### 3Ô∏è‚É£ Append (Same File Name)

```python
f = open("data.txt", "a")
f.write("This line is appended\n")
f.write("File handling complete\n")
f.close()
```

### 4Ô∏è‚É£ Read Again

```python
f = open("data.txt", "r")
print(f.read())
f.close()
```

### üß† Flow samjho (simple):

### üìÅ file handling ‚Äúwith open()‚Äù

##### ‚ùì ‚Äúwith open()‚Äù kya hota hai?

###### üëâ Normal method:


```python
file = open("data.txt", "w")
file.write("Hello")
file.close()
```

###### üëâ with open() method:

```python
with open("data.txt", "w") as file:
    file.write("Hello")
```

 ### Normal vs With open() Method

### ‚úÖ Create + Write

```python
with open("data.txt", "w") as file:
    file.write("First line\n")
    file.write("Python file handling\n")
```

### ‚úÖ Read

```python
with open("data.txt", "r") as file:
    print(file.read())
```

### ‚úÖ Append (same file)

```python
with open("data.txt", "a") as file:
    file.write("New line added\n")
```

### ‚úÖ Read again

```python
with open("data.txt", "r") as file:
    print(file.read())
```

### ‚úÖ File Pointer Control: tell() and seek()

### 1Ô∏è‚É£ tell() use karna (simple)

```python
with open("data.txt", "w") as f:
    f.write("PythonFile")

with open("data.txt", "r") as f:
    print(f.tell())     
    f.read(6)      
```

### 2Ô∏è‚É£ seek() use karna (simple)

```python
with open("data.txt", "r") as f:
    f.seek(6)          
    print(f.read())   
```

### ‚ùì JSON library kya hoti hai?

### 1Ô∏è‚É£ json import karna

### 2Ô∏è‚É£ File create + JSON write

```python
import json

data = {
    "name": "Ali",
    "age": 20,
    "city": "Karachi"
}

with open("data.json", "w") as file:
    json.dump(data, file)
```

### 3Ô∏è‚É£ JSON file read karna

```python
import json

with open("data.json", "r") as file:
    data = json.load(file)

print(data)
```

#### 1Ô∏è‚É£ List

```python
import json

data = [1, 2, 3, 4]

with open("data.json", "w") as file:
    json.dump(data, file)

with open("data.json", "r") as file:
    print(json.load(file))  
```

#### 2Ô∏è‚É£ Tuple

```python
import json

data = (10, 20, 30)

with open("data.json", "w") as file:
    json.dump(data, file)

with open("data.json", "r") as file:
    print(json.load(file))
```

#### 3Ô∏è‚É£ Dictionary

```python
import json

data = {"name": "Ali", "age": 20}

with open("data.json", "w") as file:
    json.dump(data, file)

with open("data.json", "r") as file:
    print(json.load(file))
```

#### 4Ô∏è‚É£ Nested Structure

```python
import json

data = {
    "students": ["Ali", "Sara"],
    "marks": (90, 85),
    "details": {"class": 10, "section": "A"}
}

with open("data.json", "w") as file:
    json.dump(data, file)
with open("data.json", "r") as file:
    print(json.load(file))
```
